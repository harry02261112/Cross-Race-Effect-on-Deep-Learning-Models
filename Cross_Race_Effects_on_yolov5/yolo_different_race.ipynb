```python
import os
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
conf_list_list = []

for i in range(4):
  for j in range(4):

    conf_list = []

    path = f'/content/drive/MyDrive/JHU_courses/ML_class/Final/yolov5/yolov5_new/runs/detect/race{i}_on_race{j}/labels/'

    names = [f for f in listdir(path) if isfile(join(path,f)) and f.endswith(".txt")]

    #print(names)

    for name in names:
      with open(join(path,name)) as txtfile:
        lines = txtfile.readlines()
      
      #print(lines)

      for k in range(len(lines)):

        conf_list.append(float(lines[k].split(" ")[-1].split("\n")[0]))

      conf_list_list.append(conf_list)

    print(f"Mean Race{i} on Race{j} confidence:",np.mean(conf_list))
```
